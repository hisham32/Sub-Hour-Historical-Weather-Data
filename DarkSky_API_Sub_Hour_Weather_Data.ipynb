```
import requests
import pandas as pd
```

```
# collect your api key from https://darksky.net/dev with email, it's free!! 
# collect as many api keys as you wish by registering with diffrent email address to collect large amount of data
# make a list of api keys, use string format for each key  
api_list = []

# start is the start time, should be UNIX timestamp, use https://www.unixtimestamp.com/index.php for conversion 
# maxlimit set as 1000 requests per day per key, you have to pay above that!!
# interval is the sub-hour level at which weather data is required (e.g., 5, 10, 15, 30 mins)
# lat, lon is latitude and longitude cordinates respectively in WGS 84 format 
def weather_data(start, lat, lon, interval, maxlimit=1000, api=api_list):
    
    df = pd.DataFrame()

    maxlimit = maxlimit
    start = start       
    lat = lat          
    lon = lon           
    interval = interval*60 
    
    for key in api:
        for unix in range(start, start+maxlimit, interval):
            BASE_URL = "https://api.darksky.net/forecast/"+ key +"/"+ str(lat)+","+ str(lon)+","+ str(unix) + "?exclude=flags,alerts,hourly,daily"
            response = requests.get(BASE_URL)
            data = response.json()
            df1=pd.Series(data['currently']).to_frame().transpose()
            df=df.append(df1)

        start = start+interval

    df.sort_values('time', inplace=True)
    df.reset_index(drop=True, inplace=True)
    df.to_csv('chengdu_weather.csv', index=False)

    return df
```

```
# 10-min interval weather data example of chengdu from oct-dec, 2016
# 11 keys were used in the api_list, keys are removed because of privacy reasons 
hist_data = weather_data(start=1475280000, lat=30.6528, lon=104.0681, interval=600)
hist_data
```
